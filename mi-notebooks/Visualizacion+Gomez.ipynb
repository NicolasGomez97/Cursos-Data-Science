### Consigna 2

```python
y = data["Salary"].sort_values(ascending=False)
y = y[0:100] 
y
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.hist(y,bins=10)
ax.set_title('Histograma de Salario rango mas populares')
ax.set_xlabel('Sueldos')
ax.set_ylabel('Frecuencia de catidad de personas')
```
