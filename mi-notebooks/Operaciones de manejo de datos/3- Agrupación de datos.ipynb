# Agregación de datos por categoria

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500

gender_data = []
income_date = []

for i in range(0,n):
    gender_data.append(np.random.choice(gender))
    income_date.append(np.random.choice(income))
```

```python
gender_data[0:10]
```

```python
income_date[0:10]
```

```python
height = (160 + 30 * np.random.rand(n))**1
weight = (70 + 25 * np.random.randn(n))**1
age = (50 + 25 * np.random.randn(n))**1
salary_arg = (65000 + 30000 * np.random.randn(n))**1 
```

```python
data = pd.DataFrame(
    {
        "Gender": gender_data,
        "Income": income_date,
        "Height": height,
        "Weight": weight,
        "Age": age,
        "Salary": salary_arg
    }
)
```

```python
data.head(10)
```

## Agrupacón de datos

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
double_group = data.groupby(["Gender","Income"])
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

 ### Operaciones sobre datos agrupados

```python
double_group.sum() 
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
double_group.aggregate(
    {
        "Salary": np.sum
    }
)
```

```python
double_group.aggregate(
    {
        "Age": np.mean,
        "Height": lambda h:np.mean(h)/np.std(h)
    }
)
```

### Filtrados de datos

```python
double_group.sum()
```

```python
double_group["Age"].filter(lambda x: x.sum()>2400)
```

```python
zcore = lambda x: (x-x.mean()/x.std())
```

```python
double_group.transform(zcore)
```
