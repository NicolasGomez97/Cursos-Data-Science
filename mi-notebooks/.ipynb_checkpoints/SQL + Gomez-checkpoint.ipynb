```python
class DataBase:
    import pymysql as db
    import pandas as pd
    def __init__(self):
        self.connection= db.connect(
            user='root',
            password='root',
            host='127.0.0.1',
            port=3307, 
            db= 'hr')
        self.cursor = self.connection.cursor()
        print("Conexión Exitosa")
    def select_sql(self, employeeid):
        con = self.connection
        sql = f'SELECT employee_id,first_name, last_name, email FROM employees WHERE employee_id = {employeeid}'
        return pd.read_sql_query(sql,con)
    def select_sql2(self):
        con = self.connection
        sql = 'SELECT * FROM employees'
        return pd.read_sql_query(sql,con)
database = DataBase()
```
