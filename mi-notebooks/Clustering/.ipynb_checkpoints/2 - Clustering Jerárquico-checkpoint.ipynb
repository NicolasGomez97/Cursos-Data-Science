## Clustering jerárquico y dendrograma

Antes de emepezar, pongamos un poco de notación para hablar todos el mismo idioma

* X dataset(array de n x m) de puntos a clusterizar
* n número de datos
* m número de rasgos
* Z array de enlace del cluster con la información de las uniones
* k número de cluster

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
np.random.seed(4711)
a = np.random.multivariate_normal([10,0],[[3,1],[1,4]], size = [100,])
b = np.random.multivariate_normal([0,20],[[3,1],[1,4]], size = [50,])
x = np.concatenate((a,b))
print(x.shape)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
a
```

```python
b
```

```python
Z = linkage(x, "ward")
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID")
plt.ylabel("Distancia")
dendrogram(Z,leaf_rotation=90., leaf_font_size=10.0)
```

### Respresentación gráfica de un dendrograma

```python
plt.figure(figsize=(30,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("Índice de la muestra")
plt.ylabel("Distancia")
dendrogram(Z,leaf_rotation=90., leaf_font_size=8.0, color_threshold=0.7*180)
plt.show()
```

```python
Z[-4:,]
```

### Truncar el dendrograma

```python
plt.figure(figsize=(30,10))
plt.title("Dendrograma jerárquico para el Clustering truncado")
plt.xlabel("Índice de la muestra")
plt.ylabel("Distancia")
dendrogram(Z,leaf_rotation=90., leaf_font_size=8.0, color_threshold=0.7*180, 
           truncate_mode="lastp", p=12, show_leaf_counts=False, show_contracted=True,)
plt.show()
```

### Dendograme personalizado

```python
def dendrogram_tune(*args, **kwargs):
    
    max_d=kwargs.pop("max_d", None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)
    
    ddata = dendrogram(*args,**kwargs)
        
    if not kwargs.get('no_plot', False):
        plt.title("Clustering jerárquico con Dendrograma truncado")
        plt.xlabel("Índice del Dataset (o tamaño del cluster)")
        plt.ylabel("Distancia")
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y>annotate_above:
                plt.plot(x,y,'o',c=c)
                plt.annotate('%.3g'%y, (x,y), xytext=(0,-5),
                            textcoords="offset points", va="top", ha="center")
                
    if max_d:
        plt.axhline(y=max_d, c='k')
        
    return ddata
```

```python
dendrogram_tune(Z,truncate_mode='lastp',p=12,leaf_rotation=90.,leaf_font_size=12.,show_contracted=True,annotate_above=10, max_d=20)
plt.show()
```

### Corte automático del dendrograma
inconsistency_i = (h_i-avg(h_j))/std(h_j)

```python
from scipy.cluster.hierarchy import inconsistent
```

```python
depth = 3
incons = inconsistent(Z, depth)
incons[-10:]
```

### Método del codo


```python
last = Z[-10:,2]
last_rev = last[::- 1]
print(last_rev)
idx = np.arange(1, len(last)+1)
plt.plot(idx, last_rev)

acc = np.diff(last,2)
acc_rev = acc[::-1]
plt.plot(idx[:-2]+1, acc_rev)
plt.show()
k = acc_rev.argmax() +2
print("El número óptimo de cluster es %s"%str(k))
```

```python
c = np.random.multivariate_normal([40,40],[[20,1],[1,30]], size=[200,])
d = np.random.multivariate_normal([80,80],[[30,1],[1,30]], size=[200,])
e = np.random.multivariate_normal([0,100],[[100,1],[1,100]], size=[200,])
X2 = np.concatenate((x,c,d,e),)
plt.scatter(X2[:,0], X2[:,1])
plt.show()
```

```python
Z2 = linkage(X2,"ward")
```

```python
plt.figure(figsize=(10,10))
dendrogram_tune(
    Z2,
    truncate_mode="lastp",
    p=30,
    leaf_rotation=90.,
    leaf_font_size=10.,
    show_contracted=True,
    annotate_above = 40,
    max_d = 170
)

plt.show()
```

```python
last = Z2[-10:,2]
last_rev = last[::-1]
print(last_rev)
idx = np.arange(1, len(last)+1)
plt.plot(idx, last_rev)

acc = np.diff(last,2)
acc_rev = acc[::-1]
plt.plot(idx[:-2]+1, acc_rev)
plt.show()
k = acc_rev.argmax() +2
print("El número óptimo de cluster es %s"%str(k))
```
